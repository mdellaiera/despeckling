# Process input

```python
import scipy
import numpy as np
import sys
sys.path.insert(0, '../baselines/scripts')
from baselines_utils import save_image
```

```python
KEY = 'cout'

def read_image(input_path: str) -> np.ndarray:
    """Read an image from a file."""
    data = scipy.io.loadmat(input_path)
    if KEY not in data:
        raise ValueError(f"Input file must contain '{KEY}' key.")
    return data[KEY]
```

```python
path = '../../dataset/input.mat'
data = read_image(path)

data.shape
```

```python
# split data into 1000x1000 chunks
chunk_size = 1000
chunks = []
for i in range(0, data.shape[0], chunk_size):
    for j in range(0, data.shape[1], chunk_size):
        chunk = data[i:i + chunk_size, j:j + chunk_size]
        if chunk.shape[0] == chunk_size and chunk.shape[1] == chunk_size:
            chunks.append(chunk)

print(f"Total chunks created: {len(chunks)}")
```

```python
output_dir = '../../dataset/'

for idx, chunk in enumerate(chunks):
    chunk_path = f'{output_dir}/input_{idx}.mat'
    save_image(chunk_path, chunk)
```
